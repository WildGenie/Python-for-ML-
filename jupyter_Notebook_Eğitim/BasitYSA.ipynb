```python
# Numpy kÃ¼tÃ¼phanesi iÃ§eri aktaralÄ±m
import numpy as np

# NÃ¶ron sÄ±nÄ±fÄ±mÄ±z

class Noron: 
  
  # __init__ ile baÅlangÄ±Ã§ aÄÄ±rlÄ±k ve bias deÄerlerimizi sÄ±nÄ±fÄ±mÄ±za ekliyoruz
  def __init__(self , weights , bias):
    
    self.weights  = weights
    self.bias     = bias

  # NÃ¶ron iÃ§erisinde yapÄ±lan toplam iÅlemi sonrasÄ±
  # deÄerimizi 0 ile 1 arasÄ±nda standartlaÅtÄ±rmak iÃ§in 
  # sigmoid metodumuzu oluÅturuyoruz
  def sigmoid(self,x):
    
    return 1 / (1 + np.exp(-x))
  
  def feedforward(self, data):
      
      # Girdilerin aÄÄ±rlÄ±kla ve ardÄ±ndan bias ile toplama iÅlemi
      sumResult = (np.dot(self.weights , data ) + self.bias)
      
      # En son olarak sonucu sigmoid fonksiyonundan geÃ§irerek
      # 0 ile 1 arasÄ±nda standartlaÅtÄ±rÄ±lmÄ±Å sonuÃ§ deÄerini elde ediyoruz
      return self.sigmoid(sumResult)
  
x = [4,15]       # X1 = 4 , X2 = 5
w = [0,1]        # W1 = 0 , W2 = 1
b = 0            # Bias = 0

# OluÅturduÄumuz class'Ä± kullanarak aÄÄ±rlÄ±k ve 
# bias deÄerleri ile NÃ¶ronu muzu oluÅturuyoruz
n = Noron(w ,b)

# Ä°leri besleme yÃ¶ntemi ile deÄerimizi elde ediyoruz
prediction = n.feedforward(x)

# Sonucu yazdÄ±rÄ±yoruz
print(prediction)
```

```python
# Network (AÄ) oluÅturacaÄÄ±m Python Class'nÄ± tanÄ±mlÄ±yoruz

class NeuralNetwork:
    
    def __init__(self):
        
        # AÄÄ±rlÄ±klar 0 ve 1 bias deÄeri ise 0 olarak belirliyoruz
        self.weights = [0,1]
        self.bias    = 0
        
        # 3 Adet nÃ¶ron olduÄu iÃ§in daha Ã¶nce tanÄ±mladÄ±ÄÄ±mÄ±z NÃ¶ron Class'Ä±
        # ile bu NÃ¶ronlarÄ± belirtilen AÄÄ±rlÄ±k ve Bias deÄerleri oluÅturuyoruz
        self.H1 = Noron(self.weights , self.bias)
        self.H2 = Noron(self.weights , self.bias)
        self.O1 = Noron(self.weights , self.bias)
        
        
    def feedforward(self , data):
        
        # Noron class'Ä± bize feedforwar metodu ile o nÃ¶ronun tahmin 
        # deÄerini veriyordu. H1 ve H2 nÃ¶ronlarÄ± iÃ§in girdi verileri 
        # Ã¼zerinden deÄerleri elde ediyoruz.
        
        resultH1 = self.H1.feedforward(data)
        resultH2 = self.H2.feedforward(data)
        
        
        # O1 nÃ¶ronu H1 ve H2 nÃ¶ronlarÄ± ile baÄlantÄ±lÄ±ydÄ± 
        # bu nedenle O1 nÃ¶ronuna girdi olarak H1 ve H2 
        # nÃ¶ronlarÄ±nÄ±n tahmin deÄerlerini girdi olarak iletiyoruz. 
        
        resultO1 = self.O1.feedforward([resultH1 , resultH2])
        
        
        return resultO1 
    
    

# Girdi verilerimiz Girdi KatmanÄ±nda 2 adet girdi olduÄu iÃ§in
# bu deÄerler 4 ve 15
XInput = [4 ,15]

# Netwrok objesi 
Network = NeuralNetwork()

# Girdi deÄelerinin tahmini
Prediction = Network.feedforward(XInput)

print(Prediction)
```
